**linear graph**

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
x=np.array([5,6,7])
y=np.array([1,2,4])
plt.plot(x,y,label="firstline",color='g')
plt.title("sample linear graph")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.legend()
plt.show()
```

#**scatter plot**

```python
import numpy as np
x=np.array([1,2,3])
y=np.array([4,2,5])
plt.scatter(x,y,label="scatter plot",color='g',marker="x")
plt.title("sample scatter plot")
plt.xlabel("xaxis")
plt.ylabel("yaxis")
plt.grid(True,color='k')
plt.legend()
plt.show()
```
